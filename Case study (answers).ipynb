### Question 2:
Combined with the 'Loyalty benefits' table and 'Loyalty Ranking' table, add columns '%cashback'  in 'Transactions' table and **calculate the total cashback cost in Feb 2022.** 


Cashback cost is calculated from this equation: <br>
Cashback cost = GMV * %cashback <br>
From Transactions dataset, we already have the GMV, then we have to find the responding %cashback.
From Loyalty Benefits datset, %cashback depends on the rank name and the service. We've got the service information, so we need to find the rank for the user that did the transactions. The user rank is determined by the Loyalty Points that users have earned in the last 30 days. Therefore, the furthest date that is used to calculated is Jan 3rd, 2022.

Let's create a dataframe for transactions in Feb 2022

```python
filter1 = '2022-02-01' <= Transactions['DATE']
filter2 = Transactions['DATE'] <= '2022-02-28'
Feb_transactions = Transactions[filter1 & filter2]
Feb_transactions
```

Let's have a look again at Transactions:

```python
Transactions
```

We all know that money is cashbacked for the user after they do the transaction. Also the cashback% is depend on the rank of the user and the rank depends on the loyalty points of the user **before** the transaction. However, in case there were more than one transaction happening in a day from one user, we do not know which transactions is before or after which transactions because there was no information of exact time(hour, minute, second) of the transactions. Therefore, let's assume the rank of the user will be determined on the loyalty point that the user had earned at the end of the day before the transaction. <br>
From Feb_transactions table, for each date, I will filter from the Transactions the dataset including all the transaction happen in the last 30 days and assign it with name 'reference'. The reason I do not put the following code lines available because I will put all of them under a function and take date and user as parameters. Remember our assumption? the code will be like this:

**filter3 = Transactions['DATE'] < Feb_transactions['DATE']
filter4 = Transactions[Feb_transactions['DATE'] - 30 <= Transactions['DATE']
reference_30days = Transactions[filter3 & filter4]<br>**
Next, we need to check if a user did any transaction in the last 30 days before they have been on the list of users who have transactions in Feb. If they did, I will assign all their transactions infor with name 'points' <br>
**point_calculating = reference_30days[reference_30days['User_id'] = user))<br>**
Finally, we sum all the Loyalty Points from 'points' and start over with the next date in Feb_transactions.This is the code <br>
**accumulated_points = sum(point['Loyalty Points'])** <br>
The function to calculate accumulated points for the users making money transations in Feb

```python
from datetime import timedelta
```

```python
def points_calculating(date, user):
    filter3 = date - timedelta(days=30) <= Transactions['DATE']
    filter4 = Transactions['DATE'] < date
    reference_30days = Transactions[filter3 & filter4]
    point_calculating = reference_30days[reference_30days['User_id'] == user]
    accumulated_points = sum(point_calculating['Loyalty Points'])
    return accumulated_points
```

I will extract DATE and User_id columns from Feb_transactions dataset. 

```python
Accumulated_points = Feb_transactions.loc[:,['DATE','User_id','GMV','Service Group']]
Accumulated_points
```

Now pass each row of the Date_user to the function and save the results to the Loyalty Points column :

```python
Accumulated_points['Loyalty Points'] = Accumulated_points.apply(lambda row: points_calculating(row['DATE'], row['User_id']), axis = 1)
Accumulated_points
```

Let's merge this with Ranking_rules to determine the ranking of the users for each transactions: 

```python
Ranking_name = pd.merge_asof(Accumulated_points.sort_values('Loyalty Points'),Ranking_rules.loc[:,['Class ID','Loyalty Points']], on = 'Loyalty Points')
Ranking_name
```

Then merge with Loytalty_benefits to have %cashback 

```python
Cash_back = pd.merge(Ranking_name, Loyalty_benefits, on=['Service Group', 'Class ID'])
Cash_back
```

Calculate Cachback Cost by multipling GMV with %cashback while limiting Cashback Cost of each Transaction is lower than 10,000

```python
Cash_back['Expected Cost'] = Cash_back['GMV']*Cash_back['%cashback']/100 
Cash_back['Cashback Cost'] = np.where(Cash_back['Expected Cost'] < 10000, Cash_back['Expected Cost'], 10000)
Cash_back.drop(['Expected Cost'], axis = 1, inplace = True)
Cash_back
```

**Total cashback cost in Feb 2022:**

```python
sum(Cash_back['Cashback Cost'])
```
